# 💳 Daily Transactions Analysis Project

This notebook performs end-to-end analysis of daily transaction data including data cleaning, exploratory data analysis (EDA), time series analysis, and correlation insights.

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
print('Libraries imported successfully!')
```

## 2. Load Dataset

> **Note:** Update the file path below to point to your `daily_transactions_sample.csv` file.

```python
df = pd.read_csv('daily_transactions_sample.csv')  # Update path if needed

print('First 5 rows:')
print(df.head())
print('\nShape of dataset:', df.shape)
```

## 3. Check Missing Values

```python
print('Missing Values:')
print(df.isnull().sum())
```

## 4. Data Cleaning

- Fill missing Subcategory and Note values
- Parse Date column (day-first format)
- Remove duplicates
- Ensure Amount is float type

```python
df['Subcategory'] = df['Subcategory'].fillna('Unknown')
df['Note'] = df['Note'].fillna('No Note')
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df.drop_duplicates(inplace=True)
df['Amount'] = df['Amount'].astype(float)

print('Cleaned Data Info:')
print(df.info())
```

## 5. Summary Statistics

```python
print('Summary Statistics')
df.describe()
```

## 6. Payment Mode Analysis

```python
print('Payment Mode Counts:')
print(df['Mode'].value_counts())

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Mode', order=df['Mode'].value_counts().index)
plt.title('Payment Modes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 7. Category Analysis

```python
print('Category Counts:')
print(df['Category'].value_counts())

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Category', order=df['Category'].value_counts().index)
plt.title('Transaction Categories')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 8. Top Subcategory Analysis

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Subcategory', order=df['Subcategory'].value_counts().index[:10])
plt.title('Top 10 Subcategories')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## 9. Income vs Expense

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Income/Expense')
plt.title('Income vs Expense')
plt.tight_layout()
plt.show()
```

## 10. Transaction Amount Distribution

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Amount'], bins=20, kde=True)
plt.title('Distribution of Transaction Amount')
plt.tight_layout()
plt.show()
```

## 11. Amount Distribution by Category

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Amount', y='Category')
plt.title('Amount Distribution by Category')
plt.tight_layout()
plt.show()
```

## 12. Time Series Analysis

### 12.1 Monthly Transaction Amounts

```python
monthly_data = df.resample('M', on='Date').sum(numeric_only=True)

plt.figure(figsize=(10, 5))
plt.plot(monthly_data.index, monthly_data['Amount'], marker='o', color='steelblue')
plt.title('Monthly Transaction Amount')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 12.2 Daily Transaction Amounts

```python
daily_data = df.groupby(df['Date'].dt.date).sum(numeric_only=True)

plt.figure(figsize=(10, 5))
plt.plot(daily_data.index, daily_data['Amount'], color='darkorange')
plt.title('Daily Transaction Amount')
plt.xlabel('Date')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 13. Correlation Heatmap (Category vs Amount)

```python
pivot_table = df.pivot_table(
    index='Date',
    columns='Category',
    values='Amount',
    aggfunc='sum',
    fill_value=0
)

corr_matrix = pivot_table.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

## 14. Income/Expense vs Payment Mode (Scatter)

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(data=df, x='Income/Expense', y='Mode')
plt.title('Income/Expense vs Payment Mode')
plt.tight_layout()
plt.show()
```

## 15. Final Insights

```python
print('===== FINAL INSIGHTS =====')
print('Total Transactions:', len(df))
print('Total Amount:', df['Amount'].sum())
print('Average Transaction:', round(df['Amount'].mean(), 2))

print('\nTop Category:')
print(df['Category'].value_counts().head(1))

print('\nTop Payment Mode:')
print(df['Mode'].value_counts().head(1))
```

## 16. Conclusion

- Cleaned and preprocessed daily transaction data (missing values, date parsing, duplicates).
- Analyzed payment modes, categories, subcategories, and income vs expense distribution.
- Performed time series analysis on monthly and daily transaction trends.
- Built a correlation heatmap to understand relationships between spending categories.
- Identified top transaction categories and preferred payment modes from the dataset.
